## Text Preprocessing

Nous crons une fonction pour mapper la balise POS (Part-Of-Speech) des mots (compatible avec la fonction de lemmatisation de WordNet, ce qui facilite la lemmatisation des mots) :

```python
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)
```

Nous crons ensuite une fonction pour lemmatiser les mots de la description :
- Convertit le texte en minuscules
- Supprime les balises HTML ou XML (comme "{html}")
- Supprime les URL (liens web)
- Supprime les chiffres
- Divise le texte en tokens
- Filtre les mots courts et les stopwords (mots courants)
- Filtre sur les noms et les verbes
- Lemmatise les mots en utilisant WordNet (en fonction de leur partie du discours)

```python
lemmatizer = WordNetLemmatizer()
wnl = nltk.stem.WordNetLemmatizer()
w_tokenizer = nltk.RegexpTokenizer(r'[a-zA-Z]\w+\'?\w*')

stop_words = set(nltk_stopwords.words("english"))

def preprocess(sentence):
    sentence = str(sentence)
    sentence = sentence.lower()
    sentence = sentence.replace('{html}', "")
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', sentence)
    rem_url = re.sub(r'http\S+', '', cleantext)
    rem_num = re.sub('[0-9]+', '', rem_url)
    tokenizer = RegexpTokenizer(r'[a-zA-Z]\w+\'?\w*')
    tokens = tokenizer.tokenize(rem_num)
    tagged_words = nltk.pos_tag(tokens)
    filtered_words = [w for w, pos in tagged_words if len(w) > 2 and (pos.startswith('N') or pos.startswith('V')) and w not in stop_words]
    lemma_words = [wnl.lemmatize(w, get_wordnet_pos(w)) for w in filtered_words]
    return " ".join(lemma_words)



df_category['lemmatizer_cat'] = df_category['description'].apply(preprocess)
```

```python
#fonction pour stemmer (mais nous n'utiliserons que celle pour lemmatizer car elle donne de meilleurs rsultats)
stemmer = EnglishStemmer()

def preprocess_stem(sentence):
    sentence=str(sentence)
    sentence = sentence.lower()
    sentence=sentence.replace('{html}',"") 
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', sentence)
    rem_url=re.sub(r'http\S+', '',cleantext)
    rem_num = re.sub('[0-9]+', '', rem_url)
    tokenizer = RegexpTokenizer(r'[a-zA-Z]\w+\'?\w*')
    tokens = tokenizer.tokenize(rem_num)  
    filtered_words = [w for w in tokens if len(w) > 2 if not w in stopwords]
    stem_words=[stemmer.stem(w) for w in filtered_words]
    return " ".join(stem_words)
```

### CountVectorizer

Le CountVectorizer est un outil utilis pour convertir du texte en une forme numrique, ce qui permet aux algorithmes de traitement de texte et de langage naturel de travailler efficacement avec les donnes textuelles. Voici comment il fonctionne en gnral :

- Collecte de donnes textuelles : Le CountVectorizer commence par prendre en entre un ensemble de documents textuels. Ces documents peuvent tre des articles, des critiques, des commentaires, ou tout autre type de texte.
- Tokenization : Ensuite, il divise chaque document en units plus petites, appeles "tokens". Les tokens sont gnralement des mots, mais peuvent galement tre des phrases ou d'autres units, selon la configuration.
- Comptage des occurrences : Pour chaque document, le CountVectorizer compte combien de fois chaque token apparat. Il cre un vecteur numrique pour chaque document, o chaque lment du vecteur reprsente le nombre d'occurrences d'un token spcifique.
- Cration d'une matrice : Enfin, il combine tous les vecteurs de documents pour former une matrice. Chaque ligne de cette matrice reprsente un document, tandis que chaque colonne reprsente un token unique de l'ensemble de donnes.

### TF-IDF (term frequency-inverse document frequency)

- **TF (Frquence du terme) :** Il mesure la frquence relative d'un terme dans un document donn. Cela signifie que plus un terme apparat frquemment dans un document, plus son score TF sera lev.
- **IDF (Inverse Document Frequency) :** Il mesure l'importance d'un terme dans l'ensemble du corpus. Les termes qui apparaissent frquemment dans de nombreux documents auront un score IDF faible, tandis que ceux qui sont rares auront un score IDF lev.
- **TF-IDF :** Il combine TF et IDF pour attribuer un poids  chaque terme dans chaque document. Les termes qui sont frquents dans un document mais rares dans l'ensemble du corpus auront des scores TF-IDF levs.
- **Avantages :** Le TF-IDF tient compte de l'importance relative des mots dans un document par rapport  l'ensemble du corpus, ce qui en fait une mthode plus sophistique pour la reprsentation de texte que le simple comptage de mots.

## 2nd approach: Embedding / NewGen

### Word2Vec

Word2Vec, dvelopp par Google en 2013, est une technique de modlisation de langage qui utilise des rseaux de neurones pour convertir les mots en vecteurs dans un espace smantique. Il comprend deux modles principaux :

- **CBOW (Continuous Bag of Words) :** Ce modle prdit un mot cible en se basant sur un contexte de mots environnants. Cela est ralis par l'optimisation des poids du rseau de neurones pour maximiser la probabilit du mot cible tant donn un contexte spcifique.
- **Skip-Gram :** Fonctionne de manire inverse au CBOW, en utilisant un mot cible pour prdire les mots de contexte. Ce modle est particulirement efficace pour capturer des mots avec une frquence plus faible.

Dans les deux cas, Word2Vec entrane les vecteurs de mots de manire  ce que les mots ayant des contextes similaires soient proches les uns des autres dans l'espace vectoriel, permettant ainsi de capturer leurs relations smantiques et syntaxiques. Ces vecteurs peuvent ensuite tre utiliss dans diverses applications NLP, telles que la similarit smantique, la classification de textes et d'autres tches de traitement du langage.

#### Word2Vec model creation

```python
# on dtermine la longueur moyenne des descriptions et noms de produits (en nombre de mots)
df_category['word_count'] = df_category['tokenize'].apply(len)
average_word_count = df_category['word_count'].mean()
average_word_count
```

```python
# Cration et entranement du modle Word2Vec

print("Build & train Word2Vec model ...")
w2v_model = gensim.models.Word2Vec(min_count=w2v_min_count, window=w2v_window,
                                                vector_size=w2v_size,
                                                seed=42,
                                                workers=1)
#                                                workers=multiprocessing.cpu_count())
w2v_model.build_vocab(sentences)
w2v_model.train(sentences, total_examples=w2v_model.corpus_count, epochs=w2v_epochs)
model_vectors = w2v_model.wv
w2v_words = model_vectors.index_to_key
print("Vocabulary size: %i" % len(w2v_words))
print("Word2Vec trained")
```

```python
# Prparation des sentences (tokenization)

print("Fit Tokenizer ...")
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
x_sentences = pad_sequences(tokenizer.texts_to_sequences(sentences),
                                                     maxlen=maxlen,
                                                     padding='post') 
                                                   
num_words = len(tokenizer.word_index) + 1
print("Number of unique words: %i" % num_words)
```

#### Creating the embedding matrix

```python
# Cration de la matrice d'embedding

print("Create Embedding matrix ...")
w2v_size = 300
word_index = tokenizer.word_index
vocab_size = len(word_index) + 1
embedding_matrix = np.zeros((vocab_size, w2v_size))
i=0
j=0
    
for word, idx in word_index.items():
    i +=1
    if word in w2v_words:
        j +=1
        embedding_vector = model_vectors[word]
        if embedding_vector is not None:
            embedding_matrix[idx] = model_vectors[word]
            
word_rate = np.round(j/i,4)
print("Word embedding rate : ", word_rate)
print("Embedding matrix: %s" % str(embedding_matrix.shape))
```

```python
# Cration du modle

input=Input(shape=(len(x_sentences),maxlen),dtype='float64')
word_input=Input(shape=(maxlen,),dtype='float64')  
word_embedding=Embedding(input_dim=vocab_size,
                         output_dim=w2v_size,
                         weights = [embedding_matrix],
                         input_length=maxlen)(word_input)
word_vec=GlobalAveragePooling1D()(word_embedding)  
embed_model = Model([word_input],word_vec)

embed_model.summary()
```

### BERT

BERT, dvelopp par Google AI en 2018, est un modle de traitement du langage naturel bas sur l'architecture Transformer. Contrairement  des approches antrieures comme Word2Vec, BERT est conu pour comprendre le contexte complet d'une phrase en utilisant un mcanisme d'attention bidirectionnelle, ce qui lui permet de saisir les nuances subtiles du langage.

BERT se distingue par deux caractristiques principales :

- **Attention Bidirectionnelle :**  la diffrence des modles prcdents qui lisaient le texte de manire squentielle (de gauche  droite ou de droite  gauche), BERT analyse le texte dans les deux directions simultanment. Cela lui permet de comprendre le contexte d'un mot en se basant sur tous les mots environnants, amliorant ainsi la comprhension des phrases.
- **Fine-Tuning pour Diverses Tches de NLP :** BERT est pr-entran sur un grand corpus de texte, puis peut tre finement ajust avec des donnes supplmentaires pour des tches spcifiques de NLP. Ces tches incluent la comprhension de texte, la rponse aux questions, la traduction automatique, et bien plus.

Les innovations de BERT rsident dans sa capacit  traiter efficacement le langage naturel d'une manire plus nuance et contextuellement riche, ce qui a entran des amliorations significatives dans de nombreux domaines du NLP. Son architecture flexible et sa mthode de fine-tuning le rendent adaptable  une vaste gamme d'applications de traitement du langage.

#### Fonctions

```python
# Fonction de prparation des sentences
def bert_inp_fct(sentences, bert_tokenizer, max_length) :
    input_ids=[]
    token_type_ids = []
    attention_mask=[]
    bert_inp_tot = []

    for sent in sentences:
        bert_inp = bert_tokenizer.encode_plus(sent,
                                              add_special_tokens = True,
                                              max_length = max_length,
                                              padding='max_length',
                                              return_attention_mask = True, 
                                              return_token_type_ids=True,
                                              truncation=True,
                                              return_tensors="tf")
    
        input_ids.append(bert_inp['input_ids'][0])
        token_type_ids.append(bert_inp['token_type_ids'][0])
        attention_mask.append(bert_inp['attention_mask'][0])
        bert_inp_tot.append((bert_inp['input_ids'][0], 
                             bert_inp['token_type_ids'][0], 
                             bert_inp['attention_mask'][0]))

    input_ids = np.asarray(input_ids)
    token_type_ids = np.asarray(token_type_ids)
    attention_mask = np.array(attention_mask)
    
    return input_ids, token_type_ids, attention_mask, bert_inp_tot
    

def feature_BERT_fct(model, model_type, sentences, max_length, b_size, mode='HF'):
    batch_size = b_size
    bert_tokenizer = AutoTokenizer.from_pretrained(model_type)
    time1 = time.time()

    all_last_hidden_states = []

    for step in range(0, len(sentences), batch_size):
        batch_sentences = sentences[step:step + batch_size]
        input_ids, token_type_ids, attention_mask, _ = bert_inp_fct(batch_sentences, bert_tokenizer, max_length)

        if mode == 'HF':
            outputs = model(input_ids, attention_mask=attention_mask, token_type_ids=token_type_ids)
            last_hidden_states = outputs.last_hidden_state
            all_last_hidden_states.append(last_hidden_states)

        elif mode == 'TFhub':
            text_preprocessed = {"input_word_ids": input_ids, "input_mask": attention_mask, "input_type_ids": token_type_ids}
            outputs = model(text_preprocessed)
            last_hidden_states = outputs['sequence_output']
            all_last_hidden_states.append(last_hidden_states.numpy())

    # Concatnation des tats cachs et calcul des features
    last_hidden_states_tot = np.concatenate(all_last_hidden_states, axis=0)
    features_bert = last_hidden_states_tot.mean(axis=1)

    time2 = np.round(time.time() - time1, 0)
    print("Temps de traitement : ", time2)

    return features_bert, last_hidden_states_tot
```

### USE

Universal Sentence Encoder (USE), dvelopp par Google, est un modle avanc de traitement du langage naturel conu pour encoder des phrases entires en vecteurs denses. Lanc en 2018, USE est particulirement adapt pour des tches impliquant la comprhension de la signification des phrases, telles que la classification smantique, la recherche smantique et la similarit des phrases.

USE se distingue par ses deux versions principales :

- **Version Base sur un Rseau de Neurones Transformer :** Cette version utilise l'architecture Transformer, similaire  celle de BERT, pour capturer les relations contextuelles dans les phrases. Elle est particulirement efficace pour les tches ncessitant une comprhension profonde du contexte et de la nuance dans les phrases longues.
- **Version Base sur un Rseau de Neurones Deep Averaging Network (DAN) :** La version DAN, en revanche, moyenne les embeddings de mots et passe ensuite le rsultat  travers un rseau de neurones profond. Cette approche est plus rapide et plus lgre en termes de calcul, bien qu'elle soit gnralement moins performante pour les tches ncessitant une comprhension contextuelle complexe.

L'un des principaux avantages de l'Universal Sentence Encoder est sa capacit  transformer de manire efficace et cohrente des phrases entires en vecteurs de taille fixe. Ces vecteurs peuvent tre utiliss dans diverses applications NLP, facilitant des comparaisons et des analyses smantiques prcises sur des ensembles de donnes de grande taille. USE est donc un outil puissant pour des applications telles que la recommandation de contenu, la classification de texte, et les systmes de rponse aux questions.

```python
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
```

#### USE : Features creation

```python
def feature_USE_fct(sentences, b_size) :
    batch_size = b_size
    time1 = time.time()

    for step in range(len(sentences)//batch_size) :
        idx = step*batch_size
        feat = embed(sentences[idx:idx+batch_size])

        if step ==0 :
            features = feat
        else :
            features = np.concatenate((features,feat))

    time2 = np.round(time.time() - time1,0)
    return features
```
