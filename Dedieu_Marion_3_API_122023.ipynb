```python
data = response.json()
```

Nous créons une extraction au format csv contenant les 10 premiers produits contenant pour chaque produit les données suivantes : foodId, label, category, foodContentsLabel, image.

```python
# Extraction et écriture dans un fichier CSV
with open('produits_champagne.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['foodId', 'label', 'category', 'foodContentsLabel', 'image'])

    # Parcourir les 10 premiers éléments
    for item in data['hints'][:10]:
        food = item['food']
        foodId = food['foodId']
        label = food['label']
        category = food.get('category', 'N/A')
        foodContentsLabel = food.get('foodContentsLabel', 'N/A')
        image = food.get('image', 'N/A')
        writer.writerow([foodId, label, category, foodContentsLabel, image])
```
